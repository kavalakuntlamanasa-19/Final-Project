## Attrition Count Plot

```python
sns.countplot(x='Attrition', data=df, hue='Attrition',palette='pastel')
plt.title('Employee Attrition Count', fontsize=16)
plt.xlabel('Attrition')
plt.ylabel('Count')
plt.show()
```

## Department Count Plot 

```python
plt.figure(figsize=(10, 6))
sns.countplot(y='Department', data=df, order=df['Department'].value_counts().index,hue='Department',palette='viridis')
plt.title('Number of Employees per Department', fontsize=16)
plt.xlabel('Count')
plt.ylabel('Department')
plt.show()
```

## Age Distribution Histogram

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Age'], kde=True, color='skyblue', bins=20)
plt.title('Age Distribution of Employees', fontsize=16)
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

##  Monthly Income Distribution

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['MonthlyIncome'], kde=True, color='salmon', bins=30)
plt.title('Monthly Income Distribution', fontsize=16)
plt.xlabel('Monthly Income')
plt.ylabel('Frequency')
plt.show()
```

## Violin Plot: Age by Attrition

```python
plt.figure(figsize=(8, 6))
sns.violinplot(x='Attrition', y='Age', data=df, hue='Attrition',palette='muted', split=True)
plt.title('Age Distribution by Attrition Status', fontsize=16)
plt.show()
```
